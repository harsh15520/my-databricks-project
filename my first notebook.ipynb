```python
# 1. Create a "Plan" (Transformation)
# We create a list and tell Spark to prepare it (parallelize), but it doesn't process it yet.
data = ["Hello", "Databricks", "This", "Is", "Lazy"]
rdd = spark.sparkContext.parallelize(data)

# 2. Add another "Plan" (Transformation)
# We plan to make every word uppercase. Still, no calculation happens here.
upper_rdd = rdd.map(lambda x: x.upper())

print("--- Nothing has run yet! (Lazy) ---")

# 3. Trigger the "Job" (Action)
# .collect() forces Spark to execute the plan and return the result.
result = upper_rdd.collect()

print("--- Job Triggered! Here is the result: ---")
print(result)
```

```python
# 1. Transformation (The Plan) - No work happens here
data = ["Hello", "Databricks", "This", "Is", "Lazy"]
rdd = spark.sparkContext.parallelize(data)
upper_rdd = rdd.map(lambda x: x.upper())

print("--- Nothing has run yet! (Lazy) ---")

# 2. Action (The Trigger) - This forces the work to happen
result = upper_rdd.collect()

print("--- Job Triggered! Here is the result: ---")
print(result)
```

```python
# Create sample employee data
data = [
    (1, "Alice", "Engineering", 75000),
    (2, "Bob", "Sales", 65000),
    (3, "Charlie", "Engineering", 80000),
    (4, "Diana", "HR", 60000),
    (5, "Eve", "Sales", 70000)
]

# Define column names
columns = ["id", "name", "department", "salary"]

# Create a Spark DataFrame
df = spark.createDataFrame(data, columns)

# Display the DataFrame
display(df)
```

```python
# Show DataFrame schema
df.printSchema()
display(df)
df.show()

# Count total rows
print(f"Total employees: {df.count()}")
```

```python
# Filter: Get only Engineering department employees
engineering_df = df.filter(df.department == "Engineering")

print("=== Engineering Employees ===")
display(engineering_df)
```

```python
# Transform: Add a new column with 10% bonus
from pyspark.sql.functions import col

df_with_bonus = df.withColumn("bonus", col("salary") * 0.10)

print("=== Employees with Bonus Column ===")
display(df_with_bonus)
```

```python
# Chain multiple operations (THIS IS HOW REAL PIPELINES WORK)
result = df.filter(col("salary") > 65000) \
           .withColumn("bonus", col("salary") * 0.10) \
           .withColumn("total_compensation", col("salary") + col("bonus")) \
           .select("name", "department", "salary", "total_compensation")

print("=== High Earners with Total Compensation ===")
display(result)
```

```python
from pyspark.sql.functions import col, upper

# Using chaining and col() for a clean, safe pipeline
clean_df = df \
    .filter(col("salary") > 65000) \
    .select(col("name"), col("department"),col("salary")) \
    .withColumn("name_upper", upper(col("name")))

clean_df.show()
```

```python
customers = [
    (1, "Alice"),
    (2, "Bob"),
    (3, "Cathy")
]

orders = [
    (101, 1, 250),
    (102, 1, 300),
    (103, 2, 150),
    (104, 4, 400)
]

customers_df = spark.createDataFrame(customers, ["customer_id", "name"])
orders_df = spark.createDataFrame(orders, ["order_id", "customer_id", "amount"])

result_df = (
    orders_df
    .join(customers_df, on="customer_id", how="inner")
)

result_df.show()
```

```python
from pyspark.sql import Row
transactions = [Row(id=1, item="Laptop", price=1200), Row(id=3, item="Mouse", price=25)]
trans_df = spark.createDataFrame(transactions)

# Task: Use the code below to join your 'cleaned_df' (from the previous step) 
# with this 'trans_df' on the "id" column.
joined_df = cleaned_df.join(trans_df, on="id", how="inner")
joined_df.show()
```

```python
# Join sales with employee info
# Note: using the sales_summary from previous task
sales_with_employee_info = sales_summary.join(
    emp_df, 
    sales_summary.sold_by == emp_df.name,
    "left"
)

print("=== Sales Performance with Employee Details ===")
display(sales_with_employee_info)
```

```python
# Select relevant columns and add insights
from pyspark.sql.functions import col

final_report = sales_with_employee_info.select(
    col("sold_by").alias("employee_name"),
    "department",
    "level",
    "total_sales",
    "items_sold"
).orderBy("total_sales", ascending=False)

print("=== Final Sales Report ===")
display(final_report)

# Bonus: Calculate average sales by department
dept_performance = final_report.groupBy("department") \
                               .agg(
                                   sum("total_sales").alias("dept_total_sales"),
                                   count("employee_name").alias("employee_count")
                               )

print("=== Department Performance ===")
display(dept_performance)
```

```python
# Messy sales data with missing values
messy_sales = [
    ("2024-01-19", "Laptop", 1500, "Frank"),      # Frank not in employee list
    ("2024-01-19", None, 200, "Alice"),           # Missing product name
    ("2024-01-20", "Tablet", None, "Bob"),        # Missing amount
    ("2024-01-20", "Phone", 800, None),           # Missing salesperson
    ("2024-01-21", "Monitor", 350, "Charlie")     # Clean record
]

messy_df = spark.createDataFrame(messy_sales, ["date", "product", "amount", "sold_by"])

print("=== Messy Data ===")
display(messy_df)
```

```python
from pyspark.sql.functions import col, count, when

# Count NULLs in each column
null_counts = messy_df.select([
    count(when(col(c).isNull(), c)).alias(c + "_nulls")
    for c in messy_df.columns
])

print("=== NULL Value Counts ===")
display(null_counts)
```
