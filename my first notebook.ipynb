```python
# 1. Create a "Plan" (Transformation)
# We create a list and tell Spark to prepare it (parallelize), but it doesn't process it yet.
data = ["Hello", "Databricks", "This", "Is", "Lazy"]
rdd = spark.sparkContext.parallelize(data)

# 2. Add another "Plan" (Transformation)
# We plan to make every word uppercase. Still, no calculation happens here.
upper_rdd = rdd.map(lambda x: x.upper())

print("--- Nothing has run yet! (Lazy) ---")

# 3. Trigger the "Job" (Action)
# .collect() forces Spark to execute the plan and return the result.
result = upper_rdd.collect()

print("--- Job Triggered! Here is the result: ---")
print(result)
```

```python
# 1. Transformation (The Plan) - No work happens here
data = ["Hello", "Databricks", "This", "Is", "Lazy"]
rdd = spark.sparkContext.parallelize(data)
upper_rdd = rdd.map(lambda x: x.upper())

print("--- Nothing has run yet! (Lazy) ---")

# 2. Action (The Trigger) - This forces the work to happen
result = upper_rdd.collect()

print("--- Job Triggered! Here is the result: ---")
print(result)
```

```python
# Create sample employee data
data = [
    (1, "Alice", "Engineering", 75000),
    (2, "Bob", "Sales", 65000),
    (3, "Charlie", "Engineering", 80000),
    (4, "Diana", "HR", 60000),
    (5, "Eve", "Sales", 70000)
]

# Define column names
columns = ["id", "name", "department", "salary"]

# Create a Spark DataFrame
df = spark.createDataFrame(data, columns)

# Display the DataFrame
display(df)
```

```python
# Show DataFrame schema
df.printSchema()
display(df)
df.show()

# Count total rows
print(f"Total employees: {df.count()}")
```

```python
# Filter: Get only Engineering department employees
engineering_df = df.filter(df.department == "Engineering")

print("=== Engineering Employees ===")
display(engineering_df)
```

```python
# Transform: Add a new column with 10% bonus
from pyspark.sql.functions import col

df_with_bonus = df.withColumn("bonus", col("salary") * 0.10)

print("=== Employees with Bonus Column ===")
display(df_with_bonus)
```

```python
# Chain multiple operations (THIS IS HOW REAL PIPELINES WORK)
result = df.filter(col("salary") > 65000) \
           .withColumn("bonus", col("salary") * 0.10) \
           .withColumn("total_compensation", col("salary") + col("bonus")) \
           .select("name", "department", "salary", "total_compensation")

print("=== High Earners with Total Compensation ===")
display(result)
```

```python
from pyspark.sql.functions import col, upper

# Using chaining and col() for a clean, safe pipeline
clean_df = df \
    .filter(col("salary") > 65000) \
    .select(col("name"), col("department"),col("salary")) \
    .withColumn("name_upper", upper(col("name")))

clean_df.show()
```

```python
df_from_file=spark.read.csv("/Volumes/workspace/default/my_data/sales_data.csv", header=True,inferSchema=True)
df_from_file.printSchema()
display(df_from_file)
```

```python
from pyspark.sql.functions import sum, count

# Calculate total sales per person
sales_summary = df_from_file.groupBy("sold_by") \
                            .agg(
                                sum("amount").alias("total_sales"),
                                count("product").alias("items_sold")
                            ) \
                            .orderBy("total_sales", ascending=False)

print("=== Sales Performance ===")
display(sales_summary)
```
